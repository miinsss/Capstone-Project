# Bundesland

```python
import pandas as pd
from geopy.geocoders import Nominatim
from datetime import datetime
import numpy as np
```

```python
df = pd.read_csv('contracts_with_new_data_MEAN.csv')
```

```python
# Sortiere den DataFrame nach 'ANO_SID' und 'YEAR' in aufsteigender Reihenfolge
df_sorted = df.sort_values(by=['ANO_SID', 'YEAR'])

# Entferne Duplikate, behalte die letzte Auftretung
df_unique_last = df_sorted.drop_duplicates(subset='ANO_SID', keep='last')

# Ergebnis anzeigen
print(df_unique_last)
```

```python
df_unique_last.info()
```

```python
# Sortiere den DataFrame nach 'ORPLZ', 'STRASSE' und 'YEAR' in aufsteigender Reihenfolge
df_unique_last_sorted = df_unique_last.sort_values(by=['ORTPLZ', 'STRASSE', 'YEAR'])

# Entferne Duplikate, behalte die letzte Auftretung
df_unique_last_final = df_unique_last_sorted.drop_duplicates(subset=['ORTPLZ', 'STRASSE'], keep='last')

# Ergebnis anzeigen
print(df_unique_last_final)
```

```python
df_unique_last_final.info()
```

```python
df_unique_last_final.nunique()
```

```python
plz_df = pd.read_excel("plz_df.xlsx")
```

```python
plz_df.info()
```

```python
df_bundesland_all = pd.merge(df_unique_last_final, plz_df, how='inner', left_on='ORTPLZ', right_on='PLZ')
```

```python
df_bundesland_all.info()
```

```python
df_bundesland_all.nunique()
```

```python
# Gruppieren nach der Spalte 'Bundesland'
grouped_bundesland = df_bundesland_all.groupby('Bundesland')

# Iteriere durch die Gruppen und erstelle Subsets
for bundesland, subset in grouped_bundesland:
    print(f"Subset für Bundesland: {bundesland}")
    print(subset)
```

```python

grouped_bundesland = df_bundesland_all.groupby('Bundesland')

for bundesland, subset in grouped_bundesland:
    print(f"{bundesland}: {len(subset)}")
```
