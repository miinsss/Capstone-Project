```python
import pandas as pd
from geopy.geocoders import Nominatim
from datetime import datetime
```

```python
contract_all = pd.read_csv('contract_classification.csv')
```

```python
#df = contract_all.head(100)
#print(df)
```

```python
df = contract_all
```

```python
df = df.dropna(subset=['ORTS-NAME', 'STRASSE'])
```

```python
df.isna().sum()
```

```python

geolocator = Nominatim(user_agent="CapStone")

# Funktion zur Geokodierung basierend auf der Straße und dem Ortsnamen
def get_coordinates(row):
    global count
    try:
        location = geolocator.geocode(f"{row['STRASSE']}, {row['ORTS-NAME']}, Germany")
        if location:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print(f"{timestamp}: Geokodierung erfolgreich für ANO_SID: {row['ANO_SID']} - Ortsname: {row['ORTS-NAME']} - Zahl: {count}")
            count += 1
            return (location.latitude, location.longitude)
        else:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            print(f"{timestamp}: Keine Koordinaten gefunden für ANO_SID: {row['ANO_SID']} - Ortsname: {row['ORTS-NAME']} - Zahl: {count}")
            count += 1
            return (None, None)
    except:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"{timestamp}: Fehler bei ANO_SID: {row['ANO_SID']} - Ortsname: {row['ORTS-NAME']} - Zahl: {count}")
        count += 1
        return (None, None)

# Zählvariable initialisieren
count = 0

# Anwendung der Funktion auf den DataFrame, um Longitude und Latitude zu erhalten
df['LATITUDE'], df['LONGITUDE'] = zip(*df.apply(get_coordinates, axis=1))

# Ausgabe des aktualisierten DataFrames
print(df)
```

```python
contract_with_coord = df
```

```python
contract_with_coord.to_csv('contract_with_coord.csv')
```
